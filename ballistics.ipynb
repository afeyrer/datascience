# Ballistics and the Range Equation
This is an investigation of the so-called *range equation*: $d=\frac {V^2} g \sin 2\ theta$

```python
from math import radians, sin, pi
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
def rangeformula(v, theta):
    d = v**2/9.81*sin(radians(2*theta))
    print (d)
```

```python
rangeformula (100, 10)
```

```python
angledegrees = [x/2 for x in range (0,180)]
```

```python
rangefromdegrees = [rangeformula (100, x) for x in angledegrees]
```

```python
plt.plot(angledegrees, rangefromdegrees)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (˚)')
plt.ylabel('Range (m)')
plt.show()
```

```python
rangeforangle100 = [rangeformula (100, x) for x in angledegrees]
rangeforangle200 = [rangeformula (200, x) for x in angledegrees]
plt.plot(angledegrees, rangeforangle100, label = "100 m/s")
plt.legend()
plt.plot(angledegrees, rangeforangle200, label = "200 m/s")
plt.legend()
plt.title('Range as a function of angle at various velocities')
plt.xlabel('Angle (˚)')
plt.ylabel('Range (m)')
plt.show()
```
