# Ballistics and the Range Equation
This is an investigation of the so-called *range equation*: $d=\frac {V^2} g \sin 2\ theta$

```python
from math import radians, sin, pi
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
def rangeformula(v, theta):
    d = v**2/9.81*sin((2*theta))
    return d
```

```python
angledegrees = [x/2 for x in range (0,180)]
```

```python
rangefromdegrees = [rangeformula (100, radians(x)) for x in angledegrees]
```

```python
plt.plot(angledegrees, rangefromdegrees)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (˚)')
plt.ylabel('Range (m)')
plt.show()
```

```python
rangeforangle100 = [rangeformula (100, radians(x)) for x in angledegrees]
rangeforangle200 = [rangeformula (200, radians(x)) for x in angledegrees]
plt.plot(angledegrees, rangeforangle100, label = "100 m/s")
plt.legend()
plt.plot(angledegrees, rangeforangle200, label = "200 m/s")
plt.legend()
plt.title('Range as a function of angle at various velocities')
plt.xlabel('Angle (˚)')
plt.ylabel('Range (m)')
plt.show()
```

```python
angleradians = [(x*3.14159/180) for x in angledegrees]
rangefromradians = [rangeformula (100, x) for x in angleradians]
plt.plot(angleradians, rangefromradians)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (radians)')
plt.ylabel('Range (m)')
plt.show()
```

```python
iv = range (0,201)
rangefromv = [rangeformula (x, radians(45)) for x in iv]
plt.plot(iv, rangefromv)
plt.title('Range as a function of initial velocity at 45˚')
plt.xlabel('Initial Velocity (m/s)')
plt.ylabel('Range (m)')
plt.show()
```

```python
rangefromv15 = [rangeformula (x, radians(15)) for x in iv]
rangefromv30 = [rangeformula (x, radians(30)) for x in iv]
rangefromv45 = [rangeformula (x, radians(45)) for x in iv]
rangefromv60 = [rangeformula (x, radians(60)) for x in iv]
rangefromv75 = [rangeformula (x, radians(75)) for x in iv]

plt.plot(iv, rangefromv15, label = "15˚")
plt.legend()
plt.plot(iv, rangefromv30, label = "30˚")
plt.legend()
plt.plot(iv, rangefromv45, label = "45˚")
plt.legend()
plt.plot(iv, rangefromv60, label = "60˚")
plt.legend()
plt.plot(iv, rangefromv75, label = "75˚")
plt.legend()
plt.title('Range as a function of velocity at various angles')
plt.xlabel('Initial Velocity (m/s)')
plt.ylabel('Range (m)')
plt.show()
```

**Advantages**
* cannot edit individual data points
* graph is from all data

**Disadvantages**
