```python
from numpy import radians, sin
from numpy import linspace
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
ballisticspoints = [(x,(100**2)*sin(2*radians(x))/9.81) for x in linspace(0,90,200)]
```

```python
plt.plot(ballisticspoints)
plt.show()
```

```python
df = pd.DataFrame(ballisticspoints)
```

```python
df.columns = ['Angle', 'Range']
```

```python
df
```

```python
plt.plot(df.Angle, df.Range)
plt.xlabel(df.Angle.name)
plt.ylabel(df.Range.name)
plt.legend()
plt.show()
```

```python
df2 = pd.DataFrame()
df2['Angle']=linspace(0,90,200)
df2['Range']=((100**2)*sin(2*radians(df2.Angle))/9.81)
```
