# Interest Table
This notebook calculates and graphs the value of an investment with 3.5% interest over 30 years. The formula for the value of the investment is $1000*1.

```python
from numpy import radians, sin
from numpy import linspace
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
inttable = pd.DataFrame()
inttable['years'] = range(0,31)
inttable['value_float'] = 1000*(1.035**(inttable.years))
inttable['value_str'] = ['$'+ str(round(x,2)) for x in inttable.value_float ]
```

```python
inttable
```

```python
plt.plot(inttable.years, inttable.value_float)
plt.xlabel(inttable.years.name)
plt.ylabel('value ($)')
plt.legend()
plt.show()
```
